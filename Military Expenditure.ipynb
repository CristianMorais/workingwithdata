# Analysis of military expenditures

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Military Expenditure.csv')
```

```python
df.head()
```

# Top 10 countries that have the highest military expenditures since 1980

```python
#first I drop the columns that I will not use
df2 = df.drop(df.loc[:, 'Indicator Name':'1979'], axis=1)
```

```python
#now I order the expenditures by country to get the top 10
df2['Total'] = df.loc[:, '1980':].sum(axis=1)
df2.sort_values('Total', ascending=False, inplace=True)
```

```python
df2.head()
```

```python
#let only countries 
df2 = df2[df2['Type'] == 'Country']
```

```python
#Slice only the top 10 countries 
df2 = df2.iloc[:10]
```

```python
df2.reset_index(drop=True)
```

# Visualizing the growth of the military expenditures


```python
#set the figure size for a better visualization
plt.rcParams['figure.figsize'] = [20,12]
```

```python
names = list(df2['Name'])
```

```python
for i in range(10):
    r = df2.iloc[i, 3:-1] #Do not consider the last column with the total 
    plt.plot(r)
plt.yscale('log')
plt.legend(names)
plt.title('Military Expenditures since 1980')
```

```python
plt.savefig('Military Expenditures.png')
```
